# Importing Libararies and Reading Data

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('hotel_bookings.csv')
pd.set_option('display.max_columns',32)
df.head()
```

```python
df.tail()
```

# Data Cleaning

```python
df.isna().sum()
```

```python
# Deleting unnecessary columns
df.drop(columns=['company','agent'],inplace = True)
```

```python
# We can fillna with the mode -> PRT
df['country'].fillna('PRT',inplace= True)
df['country'].isna().sum()
# we can fillna with the mode
df['children'].fillna(0,inplace= True)
df.isna().sum()
```

```python
df['reservation_status_date']= pd.to_datetime(df['reservation_status_date'])
```

```python
(df['adults'] == 0).sum()
# it doesn't sound normal so we will drop them
df = df[df['adults'] != 0]
(df['adults'] == 0).sum()
```

```python
df.describe()
```

# EDA

```python
df.describe(include = 'object')
```

```python
plt.title('Reservation Status')
plt.bar(['Not cancelled' , 'Cancelled'],df['is_canceled'].value_counts())
plt.show()
```

##### Over 30% of customers have decided to cancel their bookings, significantly affecting the hotels' revenue.

```python
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df)

# Customize legend location
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1, 1))

# Set plot title and axis labels
plt.title('Reservation status in different hotels', size=20, color='Black')
plt.xlabel('Hotel',color='Black')
plt.ylabel('Number of Reservations', color='Black')

# Customize legend labels
plt.legend(['Not Cancelled', 'Cancelled'])

# Show the plot
plt.show()
```

```python
resort_hotel= df[df['hotel']=='Resort Hotel']
City_Hotel = df[df['hotel']== 'City Hotel']
```

```python
# Calculate the mean rate in each hotel
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
City_Hotel = City_Hotel.groupby('reservation_status_date')[['adr']].mean()

plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label = 'Resort Hotel')
plt.plot(City_Hotel.index,City_Hotel['adr'],label = 'City Hotel')
plt.legend(fontsize=20)
plt.show()
```

##### The average daily rate for a city hotel is lower than that of a resort hotel, and on other days, it is even lower. Despite having fewer bookings, the resort hotel maintains a higher average rate compared to the city hotel. It is also clear that weekends and holidays tend to drive up the rates for resort hotels.

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1 = sns.countplot(x='month', hue='is_canceled', data= df)
legend_lebels,_ = ax1.get_legend_handles_labels()
plt.title('Reservation Status Per Month', size = 20)
plt.xlabel('month')
plt.ylabel('Number of Reservation')
plt.legend(['Not Canceled','Canceld'])
plt.show()
```

##### It's evident that the month of August having the highest numbers in confirmed reservations. In contrast, January has the highest number of canceled reservations.

```python
cancelled_data= df[df['is_canceled']==1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(8, 8))
plt.title('Top 10 countries with reservation canceled', color="black")
plt.pie(top_10_country, autopct='%.2f', labels=top_10_country.index)
plt.show()
```

##### So Portugal stands out as the top country with the highest number of cancellations.

```python
# We need to know the types of geusts
def Typeof(df) :
    if df['adults'] == 1 and df['children'] == 0 and df['babies'] == 0 :
        return 'Single'
    if df['adults'] == 2 and df['children'] == 0 and df['babies'] == 0 :
        return 'Couple'
    if df['adults'] > 2 and df['children'] == 0 and df['babies'] == 0 :
        return 'Group or Friends'
    else:
        return 'Family'
df['guest_type'] = df.apply(Typeof, axis=1)
df['guest_type'].value_counts().plot(kind = 'bar')
```

```python
meal =df['meal'].unique()
meal_types_full = ['Bed & Breakfast', 'Full Board', 'Half Board', 'No Meal', 'No Meal']
df['meal'].replace (meal, meal_types_full, inplace=True)
sns.countplot(data=df,x = 'meal', hue= 'guest_type')
plt.xlabel('Type of Meal')
plt.ylabel('Count of Bookings')
plt.legend(title = 'Type of Traveller', labels= ['Couples', 'Single', 'Family', 'Friends and groups']);
```
